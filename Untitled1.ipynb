```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn
import quantecon as qe
```

```python
sim_length = 10000
s = 0.6


β = 0.96

a_nsv = np.ones(sim_length)
a_sv = np.ones(sim_length)

σ_l, σ_h = 0.4, 0.8
Q = ((0.9, 0.1),
     (0.5, 0.5))

σ_mc = qe.MarkovChain(Q, (σ_l, σ_h))
σ_sim = σ_mc.simulate(sim_length)
σ_mean = σ_sim.mean()
μ = - σ_mean**2 / 2
```

```python
σ_mc.simulate(5)
```

```python
σ_mean
```

```python
ζ = randn(sim_length)
R_sv = np.exp(μ + σ_sim * ζ)
R_nsv = np.exp(μ + σ_mean * ζ)
```

```python
fig, ax = plt.subplots()

ax.plot(np.log(R_sv), label='sv')
ax.plot(np.log(R_nsv), label='nsv')
ax.legend()

plt.show()
```

```python
for t in range(sim_length - 1):
    a_sv[t+1] = β * R_sv[t+1] * s * a_sv[t] + 1    
    a_nsv[t+1] = β * R_nsv[t+1] * s * a_nsv[t] + 1
```

```python
fig, ax = plt.subplots()
ax.plot(np.arange(sim_length), a_sv, label='sv')
ax.plot(np.arange(sim_length), a_nsv, label='nsv')

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(a_sv, bins=40, alpha=0.6, label='sv')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(a_nsv, bins=40, alpha=0.6, label='nsv')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

a_vec = a_sv

# Fit a lognormal distribution for comparison and draw samples
mu = np.mean(np.log(a_vec))
sigma = np.std(np.log(a_vec))
log_obs = np.exp(mu + sigma * np.random.randn(len(a_vec)))
log_obs = - np.sort(- log_obs) # Reverse sort

a_vec = - np.sort(- a_vec) # Reverse sort
xdata = np.log(np.arange(len(a_vec)) + 1)
ydata = np.log(a_vec)
ax.plot(xdata, ydata,'bo', alpha=0.5)

lxdata = np.log(np.arange(len(log_obs)) + 1)
lydata = np.log(np.array(log_obs))
ax.plot(lxdata, lydata,'ro', alpha=0.5)

ax.set_xlabel("log rank")
ax.set_ylabel("log assets")

plt.show()
```
