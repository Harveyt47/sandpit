```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)
y1 = np.sin(x)
y2 = np.sin(5 * x)
y3 = 0.5 * np.cos(x)
y4 = 0.5 * np.cos(5 * x);

fig, axes = plt.subplots(2, 1, figsize=(9, 6))

ax = axes[0]
ax.plot(x, y1, 'b-', label="y1 label")
ax.plot(x, y2, 'g-', label="y2 label")
ax.set_ylim(-1.2, 1.2)
ax.legend()

ax = axes[1]
ax.plot(x, y3, 'r-', label="y3 label")
ax.plot(x, y4, 'y-', label="y4 label")
ax.set_ylim(-0.6, 0.6)
ax.legend()

plt.show()
```
